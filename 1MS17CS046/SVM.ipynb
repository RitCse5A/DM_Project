```python
import pandas as pd
from sklearn import svm
from sklearn import model_selection
from sklearn import metrics
from sklearn import preprocessing
import numpy as np
```

```python
data=pd.read_csv("./new_dataset5.csv")
temp=data.copy()
```

<h4>Looking at the data</h4>

```python
print(data.head())
```

```python
y=temp['Crop']
x=temp.drop('Crop',1)
```

```python
y.value_counts()
```

<h4>Crop name is a categorical attribute.We convert the categorical attribute to numerical attribute by a simple mapping.<h4>

```python
label=preprocessing.LabelEncoder()
y=label.fit_transform(y)
```

```python
np.unique(y)
```

```python
print(x.head())
```

```python
sample_size=1500
```

```python
y=y[:sample_size]
x=x[:sample_size]
```

```python
print(x.shape,y.shape)
```

<h2>Check if dataset has missing values</h2>

```python
if 'True' in x.isnull():
    print("Missing")
else:
    print("No missing data")
```

<h4>Split the dataset into test and train data<h4>

```python
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.3)
```

<h2>Normalization of data</h2>
<blockquote>The dataset has to be normalized before it is sent to any model.We can do normalization using MinMax normalization or using ZScore normalization.Both are shown</blockquote>


```python
scaler_zscore = preprocessing.StandardScaler()
scaler_minmax=preprocessing.MinMaxScaler()
```

```python
def normalize(df,choice,a):
    if choice==1:
        if a=='train':
            scaler_minmax.fit(df)
        zx = scaler_minmax.transform(df)
        return zx
    elif choice==2:
        if a=='train':
            scaler_zscore.fit(df)
        zx = scaler_zscore.transform(df)
        return zx
```

```python
x_train_zscore=normalize(x_train,2,'train')
x_test_zscore=normalize(x_test,2,'test')
x_train_minmax=normalize(x_train,1,'train')
x_test_minmax=normalize(x_test,1,'test')
```

```python
print(x_train_zscore.shape,y_train.shape)
```

```python
print(y_train)
```

<h4>Correlation matrix</h4>

<h3>Training the SVM</h3>
<blockquote>
    The gamma parameter defines how far the influence of a single training example reaches, with low values meaning ‘far’ and high values meaning ‘close’. The gamma parameters can be seen as the inverse of the radius of influence of samples selected by the model as support vectors.

The C parameter trades off correct classification of training examples against maximization of the decision function’s margin. For larger values of C, a smaller margin will be accepted if the decision function is better at classifying all training points correctly. A lower C will encourage a larger margin, therefore a simpler decision function, at the cost of training accuracy. In other words C behaves as a regularization parameter in the SVM.
</blockquote>

```python
param_grid = {'C': [0.1, 1, 10, 100, 1000],  
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
              'kernel': ['rbf','linear','poly','sigmoid']}  
  
grid = model_selection.GridSearchCV(svm.SVC(), param_grid, refit = True, verbose = 3) 
  
# fitting the model for grid search 
grid.fit(x_train_zscore, y_train) 
```

```python
print(grid.best_estimator_)
y_pred = grid.predict(x_test_zscore)
acc1=metrics.accuracy_score(y_test,y_pred)
print("Testing Accuracy for zscore normalization: ",acc1)
```

```python
grid.fit(x_train_minmax, y_train)
```

```python
print(grid.best_estimator_)
y_pred = grid.predict(x_test_minmax)
acc2=metrics.accuracy_score(y_test,y_pred)
print("Testing Accuracy for minmax normalization: ",acc2)
```

<h4>Confusion Matrix</h4>

```python
print(metrics.confusion_matrix(y_test,y_pred))
```

<h4>Classification Report</h4>

```python
print(metrics.classification_report(y_test, y_pred))
```
